```python
import matplotlib.pyplot as plt
import numpy as np

parts = ["Graphics Card", "Processor", "RAM", "Hard Drive"]
sales = [98, 110, 78, 143]

plt.bar(parts, sales)
plt.title("Computer Parts Sales")
plt.xlabel("Computer Parts")
plt.ylabel("Sales")
plt.show()
```
